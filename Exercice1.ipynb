```python
import matplotlib.pyplot as plt 
import numpy as np 
x=np.linspace(-2.5,2.5)

p=np.poly1d([1,(-2/3),-3,(8/3),0])
y=np.polyval(p,x)

p1=np.poly1d([(4/3),0,(-4/3)])
y1=np.polyval(p1,x)

y2=(1/3)*x**2+x**2-(4/3)*x

plt.plot(x,y,'#000000', label="P1")
plt.plot(x,y1,'g*', label="P2")
plt.plot(x,y2,'r.', label="P3")

plt.xlabel("absc")
plt.ylabel("ordn")
plt.legend()
plt.grid()
```
