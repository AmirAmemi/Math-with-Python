```python
# -*- coding: utf-8 -*

from pylab import *
from matplotlib.pyplot import*

x=array([44,65,71,75,87])
y=array([40,60,59,65,77])
# Affichage du nuage de points de coordonnées (X,Y)
# Nom donné à la fenêtre contenant la figure
figure("Nuage de points (version 2)")
# Titre du graphique
title('Nuage de points ')
# Légende des axes
xlabel('axe des X')
ylabel('axe des Y')
# Min et max des axes



plot(x,y,"o")
for i in range(100):
    plot(i,3.2955+i**0.83194,'b.')


grid()    # Affiche une grille
show()    # Affiche la figure	
```
